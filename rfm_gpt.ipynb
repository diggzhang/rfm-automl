```python
import pandas as pd

# ÂÅáËÆæ‰∏ä‰∏ÄÊ≠•ÁöÑÊï∞ÊçÆÊ∏ÖÊ¥óÁªìÊûúÂ≠òÊîæÂú® df ‰∏≠
df_clean = df.copy()

print(f"‚úÖ Êï∞ÊçÆÂä†ËΩΩÊàêÂäüÔºåÂÖ± {len(df_clean):,} Êù°ËÆ∞ÂΩï")


# ÂÅáËÆæÊï∞ÊçÆÂ∑≤Âä†ËΩΩ‰∏∫ df_clean
# 1Ô∏è‚É£ ËΩ¨Êç¢Êó•ÊúüÂ≠óÊÆµ
df_clean['InvoiceDate'] = pd.to_datetime(df_clean['InvoiceDate'], errors='coerce')

# 2Ô∏è‚É£ Ê†áËÆ∞ÈÄÄË¥ß
df_clean['IsReturn'] = df_clean['InvoiceNo'].astype(str).str.startswith('C')

# 3Ô∏è‚É£ Âà†Èô§Êó†ÊïàÊó•ÊúüÔºàËã•ÊúâÔºâ
df_clean = df_clean.dropna(subset=['InvoiceDate'])

# Êü•ÁúãÊó•ÊúüËåÉÂõ¥‰∏éÈÄÄË¥ßÊØî‰æã
date_range = (df_clean['InvoiceDate'].min(), df_clean['InvoiceDate'].max())
return_rate = df_clean['IsReturn'].mean()

print(f"üìÖ Êó•ÊúüËåÉÂõ¥: {date_range[0].date()} ~ {date_range[1].date()}")
print(f"‚Ü©Ô∏è ÈÄÄË¥ßËÆ∞ÂΩïÊØî‰æã: {return_rate:.2%}")

df_clean.head(3)
```

```python
# 1Ô∏è‚É£ ËÆ°ÁÆó‰∫§ÊòìÈáëÈ¢ù
df_clean['TotalPrice'] = df_clean['Quantity'] * df_clean['UnitPrice']

# 2Ô∏è‚É£ Á≠õÈô§ÈÄÄË¥ßËÆ∞ÂΩï
df_sales = df_clean[~df_clean['IsReturn']].copy()

# 3Ô∏è‚É£ ËÆ°ÁÆó RFM ÂêÑÊåáÊ†á
snapshot_date = df_sales['InvoiceDate'].max() + pd.Timedelta(days=1)

rfm = (
    df_sales.groupby('CustomerID')
    .agg({
        'InvoiceDate': lambda x: (snapshot_date - x.max()).days,  # Recency
        'InvoiceNo': 'nunique',                                   # Frequency
        'TotalPrice': 'sum'                                       # Monetary
    })
    .rename(columns={
        'InvoiceDate': 'Recency',
        'InvoiceNo': 'Frequency',
        'TotalPrice': 'Monetary'
    })
    .reset_index()
)

# Êü•ÁúãÁªüËÆ°ÊëòË¶Å
print(f"üë• ÂÆ¢Êà∑ÊÄªÊï∞: {len(rfm)}")
print("üìä RFM Ê†∑‰æã:")
rfm.head(5)
```

```python
# RFM ÊâìÂàÜ
rfm['R_Score'] = pd.qcut(rfm['Recency'], 5, labels=[5,4,3,2,1]).astype(int)
rfm['F_Score'] = pd.qcut(rfm['Frequency'].rank(method='first'), 5, labels=[1,2,3,4,5]).astype(int)
rfm['M_Score'] = pd.qcut(rfm['Monetary'], 5, labels=[1,2,3,4,5]).astype(int)

# ËÆ°ÁÆóÁªºÂêàÂàÜ
rfm['RFM_Score'] = rfm[['R_Score','F_Score','M_Score']].sum(axis=1)

# È¢ÑËßàÁªìÊûú
print("üìà RFM ÊâìÂàÜÊ†∑‰æãÔºö")
rfm.head(10)
```

```python
def segment_customer(row):
    if row['R_Score'] >= 4 and row['F_Score'] >= 4 and row['M_Score'] >= 4:
        return 'ÈáçË¶Å‰ª∑ÂÄºÂÆ¢Êà∑'
    elif row['R_Score'] >= 4 and row['F_Score'] <= 2:
        return 'ÊΩúÂäõÂÆ¢Êà∑'
    elif row['R_Score'] <= 2 and row['F_Score'] >= 3:
        return 'Ê≤âÁù°ÂÆ¢Êà∑'
    elif row['R_Score'] <= 2 and row['F_Score'] <= 2:
        return 'ÊµÅÂ§±ÂÆ¢Êà∑'
    else:
        return '‰∏ÄËà¨ÂÆ¢Êà∑'

rfm['Segment'] = rfm.apply(segment_customer, axis=1)

print("üìä ÂêÑÂÆ¢Êà∑Áæ§‰ΩìÊï∞ÈáèÔºö")
print(rfm['Segment'].value_counts())

rfm.head(10)
```

```python
print("üéØ Èò∂ÊÆµ8: Âü∫Á°ÄÊ®°ÂûãËÆ≠ÁªÉÔºàAutoGluonÁâàÔºâ")


from autogluon.tabular import TabularPredictor
import pandas as pd

print("‚Ä¢ TabularPredictor: AutoGluonË°®Ê†ºÈ¢ÑÊµãÂô®")

# ÂÅáËÆæ X_train, X_test, y_train, y_test Â∑≤ÁªèÂ≠òÂú®
# ÂêàÂπ∂ÁâπÂæÅÂíåÊ†áÁ≠æÔºåAutoGluonÈúÄË¶Å‰∏Ä‰∏™DataFrame
train_data = X_train.copy()
train_data[target_column] = y_train

test_data = X_test.copy()
test_data[target_column] = y_test

print(f"‚Ä¢ ËÆ≠ÁªÉÈõÜÂ§ßÂ∞è: {train_data.shape}")
print(f"‚Ä¢ ÊµãËØïÈõÜÂ§ßÂ∞è: {test_data.shape}")

presets_config = {
    'best_quality': {
        'ÊèèËø∞': 'ÊúÄÈ´òË¥®Èáè',
        'ÁâπÁÇπ': 'ËÆ≠ÁªÉÊó∂Èó¥ÊúÄÈïøÔºåÁ≤æÂ∫¶ÊúÄÈ´ò',
        'ÈÄÇÂêà': 'ËøΩÊ±ÇÊúÄ‰Ω≥ÊÄßËÉΩÁöÑÂú∫ÊôØ',
        'ËÆ≠ÁªÉÊó∂Èó¥': 'Âá†Â∞èÊó∂Âà∞Âá†Â§©'
    },
    'high_quality': {
        'ÊèèËø∞': 'È´òË¥®Èáè',
        'ÁâπÁÇπ': 'Âπ≥Ë°°Á≤æÂ∫¶ÂíåËÆ≠ÁªÉÊó∂Èó¥',
        'ÈÄÇÂêà': 'Áîü‰∫ßÁéØÂ¢É',
        'ËÆ≠ÁªÉÊó∂Èó¥': 'Âá†ÂçÅÂàÜÈíüÂà∞Âá†Â∞èÊó∂'
    },
    'medium_quality': {
        'ÊèèËø∞': '‰∏≠Á≠âË¥®Èáè',
        'ÁâπÁÇπ': 'Âø´ÈÄüËÆ≠ÁªÉÔºåÁ≤æÂ∫¶Â∞öÂèØ',
        'ÈÄÇÂêà': 'Âø´ÈÄüÂÆûÈ™åÂíåÂéüÂûã',
        'ËÆ≠ÁªÉÊó∂Èó¥': 'Âá†ÂàÜÈíüÂà∞Âá†ÂçÅÂàÜÈíü'
    },
    'optimize_for_deployment': {
        'ÊèèËø∞': '‰ºòÂåñÈÉ®ÁΩ≤',
        'ÁâπÁÇπ': 'Ê®°ÂûãÂ∞èÔºåÊé®ÁêÜÂø´',
        'ÈÄÇÂêà': 'Á∫ø‰∏äÈÉ®ÁΩ≤',
        'ËÆ≠ÁªÉÊó∂Èó¥': 'Âá†ÂàÜÈíü'
    },
    'ignore_text': {
        'ÊèèËø∞': 'ÂøΩÁï•ÊñáÊú¨ÁâπÂæÅ',
        'ÁâπÁÇπ': '‰∏çÂ§ÑÁêÜÊñáÊú¨ÂàóÔºåÂä†Âø´ËÆ≠ÁªÉ',
        'ÈÄÇÂêà': 'Á∫ØÊï∞ÂÄº/ÂàÜÁ±ªÊï∞ÊçÆ',
        'ËÆ≠ÁªÉÊó∂Èó¥': 'ËæÉÂø´'
    }
}

predictor = TabularPredictor(label=target_column, eval_metric='accuracy').fit(
    train_data,
    presets="high_quality"
)

print("üìã ËÆ≠ÁªÉÁªìÊûú:")
predictor.leaderboard(silent=True).head()
```

```python
# 4. Âú®ÊµãËØïÈõÜ‰∏äËøõË°åÈ¢ÑÊµã

y_pred = predictor.predict(test_data)
print("‚úÖ È¢ÑÊµãÂÆåÊàêÔºÅ")
```

```python
accuracy = (y_pred == y_test).mean()
print(f"üéØ Ê®°ÂûãÂáÜÁ°ÆÁéá: {accuracy:.3f} ({accuracy*100:.1f}%)")

if accuracy > 0.8:
    print("‚úÖ ‰ºòÁßÄÔºÅÊ®°ÂûãË°®Áé∞ÂæàÂ•Ω")
elif accuracy > 0.6:
    print("‚ö†Ô∏è ËâØÂ•ΩÔºÅÊ®°ÂûãË°®Áé∞‰∏çÈîô") 
else:
    print("‚ùå ÈúÄË¶ÅÊîπËøõÔºÅÊ®°ÂûãË°®Áé∞‰∏ç‰Ω≥")
```
